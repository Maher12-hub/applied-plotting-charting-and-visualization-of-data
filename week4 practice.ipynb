```python
np.random.seed(1234)
v1=pd.Series(np.random.normal(0,10,1000),name='v1')
v2=pd.Series(2*v1+np.random.normal(60,15,1000),name='v2')
plt.figure(figsize=(6,3))
sns.set_context('notebook')
plt.hist(v1,alpha=0.5,bins=20,label='v1');
plt.hist(v2,alpha=0.5,bins=30,label='v2');
plt.legend();
```

```python
plt.figure()
plt.hist([v1,v2],alpha=0.5,histtype='barstacked',normed=True,label=['v1','v2'])
v3=np.concatenate((v1,v2))
sns.kdeplot(v3)
plt.legend()
```

```python
plt.figure()
sns.distplot(v3,hist_kws={'color':'blue'},kde_kws={'color':'navy'})
```

```python
sns.jointplot(v1,v2,alpha=0.4,kind='hex')
```

```python
grid=sns.jointplot(v1,v2,alpha=0.4)
grid.ax_joint.set_aspect('equal')
```

```python
sns.set_style('white')
sns.jointplot(v1,v2,kind='kde',space=0)
```
